```python
val=274
for i in range(len(df1)):
    try:
        visitorId= "e1s_ficci_"+str(val)
        
        exhibitor_data = {'exhibitorAddress' : str(df1.loc[[i][0]]['exhibitorAddress']) ,
        'exhibitorBrochureURL' : '',
        'exhibitorContactNumber' : str(df1.loc[[i][0]]['exhibitorContactNumber']),
        'exhibitorContactPerson' : str(df1.loc[[i][0]]['exhibitorContactPerson']),
        'exhibitorDescription' : str(df1.loc[[i][0]]['exhibitorDescription']) ,
        'exhibitorEmail' : str(df1.loc[[i][0]]['exhibitorEmail']) ,
        'exhibitorIndustry' : str(df1.loc[[i][0]]['exhibitorIndustry']),
        #'exhibitorLogoURL' : '',
        'exhibitorName' : str(df1.loc[[i][0]]['exhibitorName']),
        'exhibitorQRCode' : str(visitorId),
        'exhibitorStallNumber' : str(df1.loc[[i][0]]['exhibitorStallNumber']),
        #'exhibitorVCUrl' : '',
        #'exhibitorVisitors' : df1.loc[[0][0]]['exhibitorVisitors'],
        'exhibitorWebsite' : str(df1.loc[[i][0]]['exhibitorWebsite']) }  
        
        visitor_data= {'visitorCityName' : str(df1.loc[[i][0]]['exhibitorAddress']) ,
        'visitorCompanyName' : str(df1.loc[[i][0]]['exhibitorName']),
        'visitorContactNumber' : str(df1.loc[[i][0]]['exhibitorContactNumber']),
        'visitorDesignation' : '',
        'visitorEmail' : str(visitorId+"@expoknot.com") ,
        'visitorId' : visitorId,
        'visitorName' : str(df1.loc[[i][0]]['exhibitorContactPerson']) }
        
        email_data= {'expoknotEmailId' : str(visitorId+"@expoknot.com") ,
        'originalEmailId' : str(df1.loc[[i][0]]['exhibitorEmail']),
        'password' : ''.join(random.choices(string.ascii_uppercase + string.digits, k=6)) }
                               
        #firebase.put("Exhibitions/"+exhibition_uid+"/exhibitionStalls/",str(i), exhibitor_data)    
        firebase.patch("Exhibitions/"+exhibition_uid+"/exhibitionStalls/"+str(i), {"exhibitorQRCode" : visitorId})    
        firebase.put("Users/Exhibitors/",val, visitor_data)    
        firebase.put("ExhibitorEmail/",val, email_data)
        
        val+=1
        
        qr = pyqrcode.create(visitorId)
        qr.png("QRs/"+str(df1.loc[[i][0]]['exhibitorName'])+".png", scale=25)
        
    except Exception as e:
        print(str(i)+"\t"+str(e))             
        
print("Upload Complete")
```
